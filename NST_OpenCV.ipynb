# **Sending The INPUT IMAGE**

```
image = cv2.imread("/content/1.jpeg")
image = imutils.resize(image, width=600)
(h, w) = image.shape[:2]
```

```
image.shape
```

```
h,w
```

```
blob = cv2.dnn.blobFromImage(image, 1.0, (w, h),
	(103.939, 116.779, 123.680), swapRB=False, crop=False)
net.setInput(blob)
output = net.forward()
```

arguments for blobFromImage(image, scaling factor, wanted reshape output, subtracts the value from respective channels B,G,R


```
output.shape
```

```
output = output.reshape((3, output.shape[2], output.shape[3]))
output[0] += 103.939
output[1] += 116.779
output[2] += 123.680
output = output.transpose(1, 2, 0)
```

```
cv2.imwrite("output_candy.jpg",output)
```
